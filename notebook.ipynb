## Exploratory Data Analysis (EDA)

### Explore ...

```python
customers_df.describe(include='all') #melihat kesimpulan parameter
```

```python
#mengabungkan orders dengan customers
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head() #menampilkan
```

```python
#grouping orders_customers_df berdasarkan city
orders_customers_df.groupby(by='customer_city').order_id.nunique().sort_values(ascending=False).reset_index().head()
```

```python
#grouping orders_customers_df berdasarkan state
orders_customers_df.groupby(by="customer_state").order_id.nunique().sort_values(ascending=False).head()
```

```python
seles_df.head() #menampilkan data teratas seles
```

```python
orders_items_df.head() #menampilkan data teratas orders
```

```python
#mengabungkan data orders_items dengan products_df
orders_items_products_df = pd.merge(
    left=orders_items_df,
    right=products_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
orders_items_products_df.head()#menampilkan
```

```python
#membuat pivot orders_items_product berdasarkan product_category_name 
orders_items_products_df.groupby(by='product_category_name').agg({
    "seller_id" : "nunique",
    "price" : "sum",
    "freight_value" : "sum"
}).sort_values(by="price", ascending=False) #mengurutkan secara descending
```

```python
#mengabungkan data orders_df dengan orders_review_df
orders_orders_review_df = pd.merge(
    left=orders_df,
    right=orders_review_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
orders_orders_review_df.head() #menampilkan
```

```python
#membuat group berdasarkan review_score tertinggi
orders_orders_review_df.groupby(by='review_score').order_id.nunique().sort_values(ascending=False)
```

```python
#mengabungkan data orders_customer dengan orders_orders_review agar dapat menjadi satu dalam all_df code selanjutnya
orders_customers_review_df = pd.merge(
    left=orders_customers_df,
    right=orders_orders_review_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
orders_customers_review_df.head() #menampilkan
```

```python
#mengabungkan data orders_customers_review_df dengan orders_items_products_df agar mendapat data review yang diinginkan
all_df = pd.merge(
    left=orders_customers_review_df,
    right=orders_items_products_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_df.head() #menampilkan
```

```python
#membuat pivot berdasarkan 'customer_state' dan 'product_category_name'
all_df.groupby(by=['customer_state','product_category_name']).agg({
    "price": "sum",
    "freight_value": "sum",
    "review_score": "mean"
})
```

```python
#membuat pivot berdasarkan'product_category_name'
all_df.groupby(by=['product_category_name']).agg({
    "price": "sum",
    "freight_value": "sum",
    "review_score": "mean"
}).sort_values(by='review_score', ascending=False) #mengurutkan secara descending
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: 

```python
#kota yang paling banyak melakukan penjualan
```

```python
#membuat grouping  customer_city dan jumlah freight_value
sum_freight_value_df = all_df.groupby(by='customer_city').freight_value.sum().sort_values(ascending=False).reset_index()
sum_freight_value_df.rename(columns={
    'freight_value' : 'quantity' #mengubah nama column 'freight_value' menjadi 'quantity'
},inplace=True) 
sum_freight_value_df.head() #menampilkan
```

```python
plt.figure(figsize=(10, 5)) #membuat canvas dengan ukuran 10,5
colors = ["#00A9FF", "#89CFF3", "#A0E9FF", "#CDF5FD","#EBF3E8"] #setting warna
sns.barplot(
    y="quantity", 
    x="customer_city",
    data=sum_freight_value_df.sort_values(by="quantity", ascending=False).head(5),
    hue=colors
) #membuat barplot megunakan seaborn
plt.title("Tingkat Penjualan berdasarkan Kota", loc="center", fontsize=15) #membuat judul
plt.ylabel(None) #menghilangkan label y
plt.xlabel(None) #menghilangkan label x
plt.tick_params(axis='x', labelsize=12) #setting ukuran x
plt.show() #menampilkan
```

### Pertanyaan 2: 

```python
#Product yang memiliki rating paling tinggi dan paling rendah
```

```python
#membuat grouping product_category_name dan mean review_score
mean_review_df = all_df.groupby(by='product_category_name').review_score.mean().sort_values(ascending=False).reset_index()
mean_review_df.head(3) #menampilkan
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 7)) #membuat figura dengan 2 column, 1 baris
 
colors = ["#00A9FF", "#89CFF3", "#A0E9FF"] #setting warna bar

#membuat barplot ke-1 menggunakan seaborn
sns.barplot(x="review_score", y="product_category_name", data=mean_review_df.head(3), hue=colors, ax=ax[0])
ax[0].set_ylabel(None) #menghilangkan label y
ax[0].set_xlabel(None) #menghilangkan label x
ax[0].set_title("Review Terbaik", loc="center", fontsize=15) #membuat sub judul
ax[0].tick_params(labelsize=15) #setting ukuran label

#membuat barplot ke-2 menggunakan seaborn
sns.barplot(x="review_score", y="product_category_name", data=mean_review_df.sort_values(by="review_score", ascending=True).head(3), hue=colors, ax=ax[1])
ax[1].set_ylabel(None) #menghilangkan label y
ax[1].set_xlabel(None) #menghilangkan label x
ax[1].invert_xaxis() #membalik barplot
ax[1].yaxis.set_label_position("right") #setting posisi label
ax[1].yaxis.tick_right() #setting posisi bar plot
ax[1].set_title("Review Terburuk", loc="center", fontsize=15) #membuat sub judul
ax[1].tick_params( labelsize=15) #setting ukuran label
 
plt.suptitle("Rata-Rata Review Berdasarkan Product Category", fontsize=20) #membuat judul utama
plt.show() #menampilkan
```
